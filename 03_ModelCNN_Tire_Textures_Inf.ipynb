# **A. Model Inference**

```python
# Import libraries
import pickle
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python
# Load models
model_aug = load_model('model_aug_ver6.keras')
```

```python
# Function to predict an image
def prediction(file, img_height=220, img_width=220):
  ## Load an image
  image_inf = tf.keras.utils.load_img(file, target_size=(img_height, img_width), color_mode='grayscale')

  ## Rescaling and reshaping image
  x = tf.keras.utils.img_to_array(image_inf)/255
  x = np.expand_dims(x, axis=0)
  image_infs = np.vstack([x])

  ## Predict
  y_pred_inf_proba = model_aug.predict(image_infs, batch_size=64)
  y_pred_inf_cls = np.argmax(y_pred_inf_proba)
  class_names = ['cracked', 'normal']
  y_pred_class_name = class_names[np.argmax(y_pred_inf_proba[0])]

  return image_inf, y_pred_class_name
```

```python
# Predict the image 1
image_inf, y_pred_class_name = prediction('tire1.jpg')

# Show the prediction
plt.imshow(image_inf)
print('Prediction : ', y_pred_class_name)
```
